```python
import networkx as nx
import matplotlib.pyplot as plt
import random
```

Criação da função 'erdos_renyi_graph(n, p):
A função recebe dois parâmetros: 'n' para o número de nós e 'p' para a probabilidade de conexão entre os nós.

Objeto G representa o grafo.

```python
def erdos_renyi_graph(n, p):
  G = nx.Graph()

  # Nós
  G.add_nodes_from(range(n)) # nós são representados por números inteiros de 0 a n-1

  # Arestas do grafo com probabilidade p

  for i in range(n): #itera sobre os nós de 0 a n-1
    for j in range(i + 1, n): # itera sobre os nós de i+1 até n-1 (garante que cada par seja considerado apenas uma vez)
      if random.random() > p: # gera um número aleatório entre 0 e 1
        G.add_edge(i, j) # se o núemro é menor que 'p', adiciona aresta entre os nós i e j

  return G
  
```

Parâmetros do grafo:
n = número de nós
p = probabilidade de conexão

```python
n = 20
p = 0.3

graph = erdos_renyi_graph(n, p) # para gerar o grafo
```

Visualização do grafo:

```python
nx.draw(graph, with_labels=True)
plt.show()
```

Aqui abordaremos outro programa utilizando o Modelo de Erdős-Rényi, mas dessa vez, com um pouco mais de complexidade

```python
import networkx as nx
import matplotlib.pyplot as plt
import random
import numpy as np
```

```python
def erdos_renyi_graph(n, p):
    G = nx.Graph()
    G.add_nodes_from(range(n))

    for i in range(n):
        for j in range(i + 1, n):
            if random.random() < p:
                G.add_edge(i, j)
    
    return G

n = 100
p = 0.15

graph = erdos_renyi_graph(n, p)
```

```python
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(graph, seed=42)
nx.draw_networkx(graph, pos=pos, with_labels=True, node_size=100, node_color='lightblue', edge_color='gray')
plt.title("Grafo aleatório - Modelo de Erdős-Rényi")
plt.axis("off")
plt.show()
```

```python
degree_sequence = sorted([d for n, d in graph.degree()], reverse=True)
degree_counts = nx.degree_histogram(graph)

for degree, count in enumerate(degree_counts):
    print(f"Grau {degree}: {count} nós")

plt.plot(degree_sequence, 'b-')
plt.xlabel('Grau')
plt.ylabel('Número de nós')
plt.title('Distribuição de Grau')
plt.show()
```

Agora, definimos uma função chamada 'calculate_degree_distribuition' que recebe um grafo como parâmetro e calcula a distribuição de grau.
Primeiro, obtemos a sequência de graus de todos os nós do grafo usando 'graph.degree()'.
Em seguida, classificamos a sequência de graus em ordem crescente e a atribuímos à variável 'degree_sequence'.
Depois, usamos a função 'np.bincount' para contar o número de ocorrências de cada grau na sequência e armazenamos os resultados em 'degree_counts'.
Por fim, calculamos a distribuição de grau dividindo 'degree_counts' pela soma de todos os valores da lista.

```python
def calculate_degree_distribution(graph):
  degree_sequence = sorted([d for n, d in graph.degree()], reverse=True)
  degree_counts = np.bincount(degree_sequence)
  degree_distribution = degree_counts / degree_counts.sum()
  return degree_distribution

degree_distribution = calculate_degree_distribution(graph)
```

Resultado:

```python
print("Distribuição de grau:")
for element in degree_distribution:
  print(element)
```

Agora, definimos uma função chamada 'calculate_average-shortest_path' que calcula o caminho mais curto médio do grafo.


Primeiro, verificamos se o grafo é conectado usando a função 'nx.is_connected'. Se o grafo for conectado, usamos a função 'nx.average_shortest_path_lenght' para calcular o caminho mais curto médio diretamente e retornamos o resultado.

Caso contrário, obtemos os componentes conectados do grafo usando 'nx.connected_components' e, para cada componente, calculamos o caminho mais curto médio no subgrafo correpondente usando 'nx.average_shortest_path_lenght'

Em seguida, somamos os caminhos mais curtos médios de cada componente e dividimos pelo número de componentes para obter uma média.

```python
def calculate_average_shortest_path(graph):
  
  if nx.is_connected(graph):
    return nx.average_shortest_path_length(graph)

  else:
    components = nx.connected_components(graph)
    average_shortest_path = 0

    for component in components:
      subgraph = graph.subgraph(component)
      average_shortest_path += nx.average_shortest_path_length(subgraph)
    
    average_shortest_path /= len(components)
    
    return average_shortest_path
    
```

```python
print("Caminho mais curto médio:", average_shortest_path)
```
